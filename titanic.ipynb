```python

import seaborn as sns
import pandas as pd

# Load the dataset
df = sns.load_dataset('titanic')

# Display the first few rows of the dataframe
df.head()
```

```python
df.info()
# Set index on the dataset

# downcast the data types
df = df.astype({col: "category" for col in df.select_dtypes(include=["object"]).columns})
df = df.astype({col: "int16" for col in df.select_dtypes(include=["int64"]).columns})
df = df.astype({col: "float32" for col in df.select_dtypes(include=["float64"]).columns})
df.info()
missing_values = df.isna().sum()
missing_values
```

```python
# Fill missing values in the Age column with the mean of the Age column
df['age'] = df['age'].fillna(df['age'].mean())
df.head()['embarked']
# Fill missing values in the Embarked column with the mode of the Embarked column
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
df.loc[:, ['embarked', 'embark_town']]


```

```python
# Find duplicate rows
duplicates = df[df.duplicated()]
print("Duplicate rows:")
print(duplicates)

# Optionally, remove duplicate rows
df = df.drop_duplicates()

# Display the DataFrame after removing duplicates
print("DataFrame after removing duplicates:")
print(df.head())
```

```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
print("Numerical columns:")
print(numerical_columns)
print(df.columns.to_list())
```
